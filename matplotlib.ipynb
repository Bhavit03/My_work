```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,4,6,7,8,8,3]
y=[12,23,13,25,36,34,45,67]
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y')
plt.plot(x,y,label='y',linestyle='dashed')
plt.legend()
plt.grid()
```

```python
bslm=[11,23,12,14,23,23,45,56,23,45,34]
bslf=[23,23,45,56,34,36,56,45,18,57,56]
plt.hist([bslm,bslf],bins=[10,20,30,40,60],rwidth=0.80,label=['men','women'])
plt.legend()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
values=[1,3,4,8,5,7,4,6]
vlabels=['as','sd','dr','rf','bg','rf','fg','vg']
plt.pie(values,labels=vlabels,explode=[0,0,0.4,0,0,0,0.6,0])
plt.axis("equal")
plt.show()
```
